```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
import matplotlib.pyplot as plt
import numpy as np

train_dataset = datasets.MNIST(root='./data', train=True, transform=transforms.ToTensor(), download=True)
train_loader = torch.utils.data.DataLoader(dataset=train_dataset, batch_size=128, shuffle=True)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

class Autoencoder(nn.Module):
    def __init__(self, latent_dim):
        super(Autoencoder, self).__init__()
        self.encoder = nn.Sequential(
            nn.Flatten(),
            nn.Linear(28*28, 256),
            nn.ReLU(),
            nn.Linear(256, latent_dim)
        )
        self.decoder = nn.Sequential(
            nn.Linear(latent_dim, 256),
            nn.ReLU(),
            nn.Linear(256, 28*28),
            nn.Sigmoid()
        )
    def forward(self, x):
        encoded = self.encoder(x)
        decoded = self.decoder(encoded)
        return decoded.view(-1, 1, 28, 28)

def train_autoencoder(latent_dim, epochs=30):
    print(f"\nTraining Autoencoder with {latent_dim} hidden units:")
    model = Autoencoder(latent_dim).to(device)
    criterion = nn.BCELoss()
    optimizer = optim.Adam(model.parameters(), lr=1e-3)

    model.train()
    for epoch in range(epochs):
        total_loss = 0
        for images, _ in train_loader:
            images = images.to(device)
            outputs = model(images)
            loss = criterion(outputs, images)
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            total_loss += loss.item()
        avg_loss = total_loss / len(train_loader)
        print(f"Epoch [{epoch+1}/{epochs}], Loss: {avg_loss:.4f}")
    return model
```

```python
autoencoder_2d = train_autoencoder(2)
```

```python
def visualize_reconstructions(model, num_images=10):
    images, _ = next(iter(train_loader))
    images = images.to(device)
    outputs = model(images[:num_images]).cpu().detach()
    fig, axes = plt.subplots(2, num_images, figsize=(num_images*2, 4))
    for i in range(num_images):
        axes[0, i].imshow(images[i].view(28,28).cpu(), cmap='gray')
        axes[0, i].axis('off')
        axes[1, i].imshow(outputs[i].view(28,28), cmap='gray')
        axes[1, i].axis('off')
    plt.show()
```

```python
visualize_reconstructions(autoencoder_2d)
```

```python
autoencoder_8d = train_autoencoder(8)
visualize_reconstructions(autoencoder_8d)
```

```python
autoencoder_32d = train_autoencoder(32)
visualize_reconstructions(autoencoder_32d)
```

```python
def sample_and_decode(model, latent_dim, num_samples=10):
    model.eval()
    with torch.no_grad():
        samples = torch.randn(num_samples, latent_dim).to(device)
        generated = model.decoder(samples).cpu()

        fig, axes = plt.subplots(1, num_samples, figsize=(num_samples*2, 2))
        for i in range(num_samples):
            axes[i].imshow(generated[i].view(28,28), cmap='gray')
            axes[i].axis('off')
        plt.show()
```

```python
sample_and_decode(autoencoder_8d, 8)
```

```python
sample_and_decode(autoencoder_32d, 32)
```
